```python
 import matplotlib.pyplot as plt
import pydarn
from datetime import datetime

# Read in Map file using SuperDARNRead, then read_map
file = "/home/eoindisney/Grd/Combined_Day1.north.map"
SDarn_read = pydarn.SuperDARNRead(file)
map_data = SDarn_read.read_map()

# Specify the datetime you want to plot
target_datetime = datetime(2023, 3, 21, 20, 10, 0)  # Example datetime, replace with your desired datetime

# Plot the map data at the specified datetime
mapplot = pydarn.Maps.plot_mapdata(map_data, datetime=target_datetime, lowlat = (50), imf_dial = False)

plt.show()
```

```python
import matplotlib.pyplot as plt 
import datetime as dt
import pydarn

#Read in GRID file for Day 0
stime = dt.datetime(2023, 3, 20, 23, 30)
grid_file_0 = "/home/eoindisney/Grd/Combined_Day0.grd"
grid_data_0 = pydarn.SuperDARNRead(grid_file_0).read_grid()

pydarn.Grid.plot_grid(grid_data_0, start_time=stime,
                      colorbar_label='Velocity (m/s)',
                      radar_label=True, line_color='blue',
                      fov_color='grey', lowlat = (50))

plt.show()
```
